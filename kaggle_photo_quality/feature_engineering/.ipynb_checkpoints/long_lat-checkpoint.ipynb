I just want to engineer good features for longitude and latitude here. It's the classic case of combining 2 variables into 1. I can judge my performance based on the linear regression between what I have with y variable.

Some ideas:
1. standard scaling
2. PCA
3. k-means 
4. convert to angular: x = xcos + ysin, y = ycos - xsin

```python
import numpy as np 
import pandas as pd 
```

```python
# just need a quick plotting tool
%matplotlib inline
from matplotlib import pyplot as plt

# df needs to hv a latitude and longitude field
def plot(df):
    plt.figure(figsize=(8,6))
    plt.scatter(df['longitude'], df['latitude'], color='green', label='input scale', alpha=0.5)
    plt.grid()
    plt.show()
```

```python
# measure of how good the features are
from sklearn.linear_model import LogisticRegression
from sklearn.cross_validation import train_test_split
from sklearn.metrics import accuracy_score

def test(X, y):
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
    lr = LogisticRegression()
    lr.fit(X_train, y_train)
    return accuracy_score(y_test, lr.predict(X_test))
```

```python
train = pd.read_csv("../data/training.csv")
plot(train)
print test(train[["latitude", "longitude"]], train.good)

concise = train[["latitude", "longitude", "good"]]
```

**1. Standard scaling**

I read scikit learn's documentation: http://scikit-learn.org/stable/modules/preprocessing.html
Im confused about whats the difference between scaling and normalizing.
Ok, this blog explains it pretty well: http://sebastianraschka.com/Articles/2014_about_feature_scaling.html#z-score-standardization-or-min-max-scaling
Normalizing is like softmax, u want the numbers to be divided by the sum or rms. Scaling or standardization is like minus mean divide by sd.

Oh also, interesting that theres a l1 vs l2 norm for the normalization package. 
https://rorasa.wordpress.com/2012/05/13/l0-norm-l1-norm-l2-norm-l-infinity-norm/

```python
from sklearn import preprocessing

zscaler = preprocessing.StandardScaler().fit(concise[['latitude', 'longitude']])
concise[['latitude', 'longitude']] = zscaler.transform(concise[['latitude', 'longitude']])
print test(concise[["latitude", "longitude"]], concise.good)
plot(concise)
```

**2. PCA**

Standard PCA, take all eigen values, but just normalizing the features

RIP, same accuracy :(, but I do hope that this is a better feature

```python
from sklearn.decomposition import PCA

pca = PCA()
# concise[["lat_pca", "long_pca"]] = 
pca_results = pca.fit_transform(concise[["latitude", "longitude"]])
pca_concise = pd.DataFrame()
pca_concise["latitude"] = pca_results[:,0]
pca_concise["longitude"] = pca_results[:,1]
plot(pca_concise) # wow this is a pretty different shape =)
print test(pca_concise[["latitude", "longitude"]], concise.good) 
```

**3. K-Means**

I expect accuracy to drop... Let's see...

```python
from sklearn.cluster import KMeans, AgglomerativeClustering
from sklearn.preprocessing import OneHotEncoder

cluster_dists = KMeans(n_clusters=8).fit_transform(concise[["latitude", "longitude"]])
print test(cluster_dists, concise.good)

cluster_dists2 = KMeans(n_clusters=24).fit_transform(concise[["latitude", "longitude"]])
print test(cluster_dists2, concise.good)

km_clusters = KMeans(n_clusters=24).fit_predict(concise[["latitude", "longitude"]])
km_clusters = km_clusters.reshape(km_clusters.shape[0], 1)
km_clusters = OneHotEncoder().fit_transform(km_clusters)
print test(km_clusters, concise.good)

# h_clusters = AgglomerativeClustering(n_clusters=8).fit_transform(concise[["latitude", "longitude"]])
# print h_clusters.shape # ok too much data for pca >.>
```
